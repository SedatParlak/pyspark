# Spark DataFrame Basic Operations

**Imports**

```python
import findspark
findspark.init('/home/sedat/spark-3.3.2-bin-hadoop3')
from pyspark.sql import SparkSession
```

**Creating SparkSession and read csv**

```python
spark = SparkSession.builder.appName('ops').getOrCreate()
```

```python
df = spark.read.csv('appl_stock.csv', inferSchema=True, header=True)
```

```python
df.printSchema()
```

```python
df.show()
```

```python
df.head(2)[0]
```

**Total data**

```python
df.count()
```

**Filtering data**

```python
df.filter('Close < 500').show()
```

```python
df.filter('Close < 500').select('Open').show()
# df.filter(df['Close'] < 500).select('Open').show() -----> Gives same results
```

```python
df.filter('Close < 500').select(['Open', 'Close', 'Volume']).show()
# df.filter(df['Close'] < 500).select(['Open', 'Close', 'Volume']).show() -----> Gives same results
```

```python
df.filter('Close < 500 AND Open > 250').select(['Open', 'Close', 'Volume']).show()
# df.filter((df['Close'] < 500) & (df['Open'] > 250)).select(['Open', 'Close', 'Volume']).show()
```

```python
df.filter(df['Low'] == 197.16).show()
```

```python
result = df.filter(df['Low'] == 197.16).collect()
```
